## 8. Übung: Confusion Matrix interpretieren

Diese Beispielwerte dienen nur zur Interpretation. Es wird noch kein eigenes Modell trainiert.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_curve,
    roc_auc_score
)

# 1. Zielvariable prüfen
print(df["credit_risk"].value_counts())
# 2. Daten vorbereiten
df_ml = df.copy()
# Zielvariable separat speichern
y = df_ml["credit_risk"]
# Merkmale ohne Zielvariable
X = df_ml.drop(columns=["credit_risk"])
# Kategoriale Variablen umwandeln
X = pd.get_dummies(X, drop_first=True)
# Falls Zielvariable Text enthält, in 0/1 umwandeln
if y.dtype == "object":
    y = y.map({"bad": 0, "good": 1})

print("X-Form:", X.shape)
print("y-Verteilung:")
print(y.value_counts())
# 3. Trainings- und Testdaten aufteilen
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
# 4. Modell trainieren
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    random_state=42
)

rf.fit(X_train, y_train)
# 5. Vorhersagen

y_pred = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]


# 6. Bewertung

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nKlassifikationsbericht:")
print(classification_report(y_test, y_pred))
# 7. Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=["Schlechtes Risiko", "Gutes Risiko"]
)

disp.plot(cmap="Blues")
plt.title("Confusion Matrix – German Credit")
plt.show()

print("Confusion Matrix:")
print(cm)

```

### Reflexion Confusion Matrix

## 9. Übung: ROC-Kurve verstehen

```python

# 8. ROC-Kurve
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
auc = roc_auc_score(y_test, y_proba)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f"ROC-Kurve (AUC = {auc:.2f})")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray", label="Zufall")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC-Kurve – German Credit")
plt.legend()
plt.grid(True)
plt.show()

print(f"AUC: {auc:.2f}")
```
